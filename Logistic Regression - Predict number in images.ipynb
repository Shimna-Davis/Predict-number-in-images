```python
# Import necessary libraries and modules

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np           # for numerical operations
from sklearn import metrics  #  for evaluating model performance
%matplotlib inline           # for displaying plots inline
```

```python
# Load the Digits dataset and print the shapes of the data and target arrays

digits = load_digits()          # loads the digits dataset, which consists of hand-written digit images
print('Image Data Shape', digits.data.shape)
print('Label Data Shape', digits.target.shape)
```

```python
# Visualize the first five images from the dataset

plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1,5,index+1)          # plt.subplot - The images are displayed side by side
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize=20)
```

```python
# Split the dataset into training and testing sets

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data, digits.target, test_size=0.23, random_state=2)
# x_train and y_train are the training data and labels, and x_test and y_test are the testing data and labels
```

```python
print('Shape ')
print('x_train :',x_train.shape)
print('x_test :',x_test.shape)
print('y_train :',y_train.shape)
print('y_test :',y_test.shape)
```

```python
# Create and train a Logistic Regression classifier

from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train, y_train)
# fit - A Logistic Regression classifier is initialized and trained on the training data
```

```python
# Predict using the trained model

logisticRegr.predict(x_test[0].reshape(1,-1))
```

```python
logisticRegr.predict(x_test)
```

```python
# Calculate the accuracy score of the model

predictions = logisticRegr.predict(x_test)

# The trained model is used to predict the labels for the test data.
```

```python
score = logisticRegr.score(x_test, y_test)
score
 
# score is calculated using the score method, which computes the accuracy of the model on the test data.
```

```python
# Compute the confusion matrix

from sklearn.metrics import confusion_matrix   
# confusion matrix is computed using the confusion_matrix function from the metrics module.
# Confusion matrix shows the count of true +ve, true -ve, false +ve, false -ve predictions
cm = metrics.confusion_matrix(y_test, predictions)
cm
```

```python
# Visualize the confusion matrix as a heatmap

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=4, square=True, cmap="jet")

# A heatmap visualization of the confusion matrix is created using Seaborn's heatmap function. 
# The annotations inside each cell show the percentage of correct predictions. 
# The labels and title are added to the heatmap

plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = ("Accuracy Score: ",format(score))
plt.title(all_sample_title, size=10)
```

```python
# Confusion_matrix - identifing the accuracy of the model
# the diagonal should have maximum numbers
# the cells otherthan diagonal have very few numbers

# in 1st row - the number fed =0, number predicted=0
# so the predicted and actual values are same

 #in 1st row '37' = 37 numbers are predicted as the number = 0,
 #and 4 is predicted in 1 time ==> misclassifications[inaccurately predicted]

#for an ideal model, every number should be in diagonal and rest cells are '0', thus accuracy is 100
```

```python
# Visualize a few examples of correctly classified and misclassified digits

index=0
classifiedIndex = []
for predict, actual in zip(predictions, y_test):
    if predict==actual:
        classifiedIndex.append(index)
    index==1
plt.figure(figsize=(20,3))
for plotIndex, wrong in enumerate(classifiedIndex[0:4]):
    plt.subplot(1, 4, plotIndex+1)
    plt.imshow(np.reshape(x_test[wrong], (8,8)), cmap=plt.cm.gray)
    plt.title('Predicted: {}, Actual: {}' .format(predictions[wrong], y_test[wrong]), fontsize=20)
    
# A heatmap visualization of the confusion matrix is created using Seaborn's heatmap function. 
# The annotations inside each cell show the percentage of correct predictions. 
# The labels and title are added to the heatmap
```
